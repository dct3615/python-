等额本息法：

计算公式：$月还款额=本金*月利率*[(1+月利率)^n/[(1+月利率)^n-1]$

式中n表示贷款月数，^n表示n次方，如^240，表示240次方(贷款20年、240个月)
月利率=年利率/12

总利息=月还款额*贷款月数-本金

```python
# -*- coding: utf-8 -*-
while True:
    try:
        loanAmount = 10000 * float(input("请输入贷款金额，单位（万）："))
        break
    except:
        print("输入错误，请重新输入贷款金额！")
while True:
    try:
        annualRate = float(input("请输入年利率,单位（%）："))/100
        break
    except:
        print("输入错误，请重新输入年利率！")    
while True:
    try:
        loanYear = int(input("请输入贷款年限："))
        break
    except:
        print("输入错误，请重新输入贷款年限！")        
monthRate = annualRate/12
n = 12 * loanYear
monthRefund = loanAmount * monthRate *(pow((1 + monthRate),n)/(pow((1 + monthRate),n) - 1))
totalInrerest = monthRefund * n - loanAmount
print("月还款金额为%.2f，总利息为%.2f："%(monthRefund,totalInrerest))
```

等额本金

每月还款金额 (简称每月本息) =（贷款本金 / 还款月数） + (本金 - 已归还本金累计额) X 每月利率。

```python
# -*- coding: utf-8 -*-
while True:
    try:
        loanAmount = 10000 * float(input("请输入贷款金额，单位（万）："))
        break
    except:
        print("输入错误，请重新输入贷款金额！")
while True:
    try:
        annualRate = float(input("请输入年利率,单位（%）："))/100
        break
    except:
        print("输入错误，请重新输入年利率！")    
while True:
    try:
        loanYear = int(input("请输入贷款年限："))
        break
    except:
        print("输入错误，请重新输入贷款年限！")  

n = 12 * loanYear
monthRate = annualRate/12
refundMoney = 0
totalInrerest = 0

for i in range(n): 
    monthRefund = loanAmount/n + (loanAmount - refundMoney)*monthRate
    totalInrerest += (loanAmount - refundMoney)*monthRate
    refundMoney += monthRefund
    print('第%d月还款的金额为%.2f'%(i+1,monthRefund))
print('总利息为%.2f'%totalInrerest)
```
