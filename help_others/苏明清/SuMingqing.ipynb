```python
import numpy as np

Ex = np.array([30,65,75,85,95])
En = np.array([25.44,4.24,4.24,4.24,4.24])
He = 0.01
n = 5000
```

```python
Ex2 = 81.5
En2 = 2.34
He2 = 0.69
x2 = []
y2 = []
Enn2 = []

Enn2 = np.random.randn(n,1) * He2 + En2
x2 = Enn2 * np.random.randn(n,1) + Ex2
y2 = np.exp(-(x2 - Ex2)**2/(2*Enn2**2))
```

```python
Ex3 = 84.6416
En3 = 3.1212
He3 = 1.1026
x3 = []
y3 = []
Enn3 = []

Enn3 = np.random.randn(n,1) * He3 + En3
x3 = Enn3 * np.random.randn(n,1) + Ex3
y3 = np.exp(-(x3 - Ex3)**2/(2*Enn3**2))
```

```python
Enn = np.random.randn(n,1) * He + En
x = Enn * np.random.randn(n,1) + Ex
y = np.exp(-(x - Ex)**2/(2*Enn**2))
```

```python
# -*- coding=utf-8 -*-
import matplotlib.pyplot as plt
import numpy as np
from pylab import * # 中文系显示
mpl.rcParams['font.sans-serif'] = ['SimHei']

plt.figure(figsize=(12,6), dpi= 100)  # 宽20，高8，每英寸上像素点的个数

for i in range(x.shape[1]):
    plt.scatter(x[:,i],y[:,i],s = 0.1,color = 'b')
# plt.scatter(x2,y2,s = 0.1,color = 'r')
plt.scatter(x3,y3,s = 0.1,color = 'y')

# 设置横轴间隔
plt.xticks(range(0,101,10))
plt.xlim(40,100)
#设置坐标轴名称
plt.xlabel('评价值')
plt.ylabel('隶属度')
plt.grid()  # 生成网格
# 保存
#plt.savefig('./picture/t1.png')
plt.show()
```
